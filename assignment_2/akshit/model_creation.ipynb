```python
import numpy as np
import os
```

```python
import torch
```

```python
data = torch.zeros([210,12])
```

```python
values = open('data.txt',"r")
i=0
for line in values:
    
    j=0
    splitted_line = line.split(' ')
    splitted_line = splitted_line[:-1]
    
    for numbers in splitted_line:
        number = int(numbers)
        data[i,j] = number
        j+=1
    i+=1
```

```python
data.dtype
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(5,1, bias=False)
        
    def forward(self,x):
        x = (self.fc1(x))
        return x
    
    
model = Model()    
```

```python
import torch.optim as optim
```

```python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(30):
    running_loss = 0.0
    
    for j in range(10):
        a = torch.zeros((210,5))
        q = 1
        k=j
        while k+2>0 and q<6:
            a[:,-q] = data[:,k+1]
            k-=1
            q+=1

        optimizer.zero_grad()
        output = model(a)
        out = torch.zeros((210,1))
        out[:,0] = data[:,j+2]


        loss = criterion(output, out)
        loss.backward()
        torch.nn.utils.clip_grad_norm(model.parameters(),5.0)
        optimizer.step()
        print(a)
        print(out)
        print(output)
        
        running_loss += loss
        if j==9:
            print(' %.3f' %( running_loss))
            running_loss = 0.0
            
print((abs((output-out)/out)*100).mean())            
print('finished training')                
                
```
